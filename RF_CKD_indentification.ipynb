```python
import pandas as pd
```

```python
dataset = pd.read_csv("CKD.csv")
dataset
```

```python
dataset.shape
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.shape
```

```python
I = dataset [['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
D = dataset [['classification_yes']]
```

```python
I.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(I,D,test_size=1/3,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

grid_param = {'criterion' : ['gini', 'entropy', 'log_loss'], 
              'max_features' : ['sqrt', 'log2', None]}

grid = GridSearchCV(RandomForestClassifier(), grid_param,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
y_pred = grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
class_report = classification_report(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test,y_pred,average='weighted')
```

```python
re = grid.cv_results_
print ("Best param {} and best score {}".format(grid.best_params_,grid.best_score_))
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
